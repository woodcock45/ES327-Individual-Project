```python
# Calculation

N_A = 6.02*1e+23 # avogadros number
M = (Co*1e-6)/Mo # molarity (mol/ml)
mol = M*V # total number of moles
##Nm = mol*N_A
Nm = mol*N_A

print ("Number of available molecules:" + format(Nm,"10.2E"))
```
